```python
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import pylab
```

```python
x, y = np.loadtxt('CH4_mbar1000_T300',unpack=True)
xa = x[:7000]
ya= y[:7000]
plt.plot(xa,ya,'y',markersize=0,label='CH4')
plt.xlabel('Wavenumber (cm-1)')
plt.ylabel('Opacity')
#plt.title('Opacity spectra CH4')

#plt.savefig('Opacity spectra CH4')

x3, y3 = np.loadtxt('../radtran/out/transmission/ISOPREN_5T.TXT', unpack=True)
#density of isoprene: 2.479e13 [1/cm^3]
#path length = 19.94 cm
y5 = (y3*np.log(10))/(6.022140857*10**23*2.805*100*10**-6/68.12)
#x3b=x3[45000:]
#y5b=y5[45000:]
plt.plot(x3,y5,'k',markersize=0,label='Isoprene')
plt.xlabel('Wavenumber (cm-1)')
#plt.ylabel('Intensity')
#plt.title('PNNL data')
#plt.savefig('Isoprene normal spectra')
plt.legend()
plt.show()

plt.plot(xa,np.log10(ya),'y',markersize=0,label='CH4')


plt.plot(x3,np.log10(y5),'k',markersize=0,label='Isoprene')
plt.xlabel('Wavenumber (cm-1)')
#plt.ylabel('Intensity')
#plt.title('PNNL data')
#plt.savefig('Isoprene normal spectra')
plt.legend()
plt.show()

```
