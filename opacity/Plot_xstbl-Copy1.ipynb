```python
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import pylab
```

```python
x, y = np.loadtxt('CH4_mbar01_T200',unpack=True)
plt.plot(x,np.log10(y),'y',markersize=0,label='0.1 mbar')
plt.xlabel('Wavenumber (cm-1)')
plt.ylabel('Opacity')

x2, y2 = np.loadtxt('CH4_mbar1_T200', unpack=True)
plt.plot(x2,np.log10(y2),'g',markersize=0,label='1 mbar')
x3, y3 = np.loadtxt('CH4_mbar10_T200', unpack=True)
plt.plot(x3,np.log10(y3),'b',markersize=0,label='10 mbar')
x4, y4 = np.loadtxt('CH4_mbar100_T200', unpack=True)
plt.plot(x4,np.log10(y4),'r',markersize=0,label='100 mbar')
x5, y5 = np.loadtxt('CH4_mbar1000_T200', unpack=True)
plt.plot(x5,np.log10(y5),'m',markersize=0,label='1000 mbar')
plt.xlabel('Wavenumber (cm-1)')
plt.legend()
plt.title('CH4')
plt.show()

```
